# Data Cleaning

```python
df = pd.read_excel(r"/content/Customer Call List.xlsx")
```

```python
df
```

Dropping duplicate data and columns we don't necessarily need

```python
df = df.drop_duplicates()
```

```python
df = df.drop(columns = "Not_Useful_Column")
```

Clean the Last_Name Column to remove special characters and whitespaces

```python
import re

pattern = r'[^a-zA-Z]'
df["Last_Name"] = df["Last_Name"].str.replace(pattern, '', regex=True)

# or use strip
# df["Last_Name"].str.strip("123._/")
```

Remove special characters from numbers to reformat it

```python
pattern_num = r'[^0-9]'
df["Phone_Number"] = df["Phone_Number"].str.replace(pattern_num, '', regex=True)
```

add the strip '-' symbol on the numbers so it have the same format

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:])
```

```python
df
```

remove the "--" values on Phone_Number column

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('--', '')
df
```

```python
df.isna().sum()
```

split the Address column

```python
df[['Street', 'State', 'Zip_Code']] = df['Address'].str.split(',', 2, expand=True)
```

```python
df
```

Re-label the Paying Customer column, fill the blank values, and replace N/a values in the dataframe

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df
```

```python
df.fillna('', inplace=True)
```

```python
df
```

```python
df = df.replace('N/a', '')
df
```

Relabel the Do_Not_Contact column and remove the rows with blank values

```python
df = df[~df['Do_Not_Contact'].isin(['Yes', 'Y'])]
```

```python
df
```

```python
df['Do_Not_Contact'] = 'N'
df
```

```python
df = df.loc[df['Phone_Number'] != '', :]
df
```

```python
df = df.reset_index(drop=True)
df
```
